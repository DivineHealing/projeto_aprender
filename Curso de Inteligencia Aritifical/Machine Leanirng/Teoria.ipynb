<h3><strong>Machine Learning</strong></h3>

<h4>Classificações</h4>
No Machine Learning, ou aprendizado de máquina, nós temos algumas classificações sendo elas: 

+ Naive bayes
    - Trabalha com probabilidades, fazendo a verificação com modelo de probabilidade e calculando riscos com base no historico apresentando.
+ Árvore de decisão
    - Ele irá analisar os dados da base de dados historica, então irá gerar uma arvore de decisão baseada nela, onde os parametros e decisoes do algoritmo serão definidos pelos nós folhas criados.
+ Aprendizagem por Regras
    - O Algoritmo realiza alguns calculos estatisticos, retornando regras baseadas nos dados historicos, após os calculos.
+ Aprendizagem Baseada em Instâncias
    - Baseado no calculo de Distancia Euclidiana
+ SVM
    - O algoritmo irá traçar uma linha para separar os dados, buscando traçar essa linha com as margens máximas dos vetores, baseado no historico de dados apresentado.
    - Antes do Deep Learning, erá considerado o melhor algoritmo.
+ Regressão Logística
    - O algoritmo irá procurar realizar previsões futuristicas relacionada aos dados inseridos, classificando eles.

<h4>Regressão</h4>

A Regressão busca prever numeros, baseado nos dados inseridos, como por exemplo: Gastos com propaganda(x) -> **valor de venda(y)**<br>
Nessa situação, você busca prever o valor de venda baseado com os valores de propaganda feitos.<br>
+ **X** Atributos previsores
+ **Y** Atributo de Retorno

**Regressão linear**<br>
Serve para realizar previsões futuristicas.
