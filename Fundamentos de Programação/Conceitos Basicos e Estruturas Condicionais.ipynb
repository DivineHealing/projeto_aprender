<h5> Exemplo:</h5>

```python
A = 2
B= 3
Resultado = 0

Resultado = A + B
print(f'Adição: {Resultado}')
Resultado = B - A
print(f'Subtração: {Resultado}')
Resultado = A * B
print(f'Multiplicação: {Resultado}')
Resultado = (A + 4)/3
print(f'Prioridade: {Resultado}')
```

<h3>Manipulação de Texto - Strings</h3>

Dados do tipo texto, sequência de caracteres alfanuméricos
+ **Tipo** str
+ **Função** input()
    + **Retorna um dado do tipo** str
        + **funções úteis p/conversão:** int() e float()
+ **Operações úteis sobre** str
    + **Função** len(<'string>)
    + **Concatenação** operador +
    + **Busca** if <'substring> in <'string>
    + **Métodos** lower() e upper()

Veja o exemplo abaixo:<br>
```python
nome = input("digite seu primeiro nome: ")
sobrenome = input("digite seu Sobrenome: ")

tamanho = len(nome) # Retorna a quantia de caracteres
print(f"O comprimento do nome é {tamanho}")

nomeComp = nome + " " + sobrenome
print(f"Apos concatenar as strings temos {nome}")

if sobrenome in nome:
    print(f"O sobrenome {sobrenome} está contido no nome {nomeComp}.")

print(f"O nome completo em minusculo é {nomeComp.lower()}")
print(f"O nome completo em maiuscuto é {nomeComp.upper()}")
```

<h5>Introdução</h5>

Há casos onde é necessario definir caminhos diferentes de acordo com situações especifícas, por exemplo.
+ Tomar decisões;
+ Escolher entre diferentes opções;
+ Realizar comparações.

Imagine um caso em que é preciso determinar um valor, informado pelo usuário, é um número negativo?
+ dependendo dessa resposta, o algoritmo irá executar um bloco de comandos ou simplesmente ignorá-lo.

Desenvolver um programa que peça para o **usuário informar dois números**. O programa deverá informar o seguinte:
+ a) qual é o maior numero?
+ b) Qual é o menor número?
+ c) Os números são iguais?

```python
a = int(input("Insira um numero: "))
b = int(input("Insira um novo numero: "))

if (a > b):
    print("O primeiro numero é o maior!")
    print("O segundo numero é o menor!")
    
elif (b > a):
    print("O segundo numero é o maior!")
    print("O primeiro numero é o menor!")

if (a == b):
    print("Os numeros são iguais!")
```

<b>Estrutura condicional ou estrutura de decisão</b>

+ A sequência de passos a ser executada depende de uma condição;
+ Caso a condição seja verdadeira, o bloco de comandos é executado;
+ Caso a condição  não seja verdadeira,  o bloco de comandos é ignorado;
+ Permite interromper e/ou redirecionar o fluxo de um algoritmo.

As estruturas condicionais são usadas para **"selecionar um caminho"** que será executado, baseado em uma condição. A **condição** deve ser composta por de uma ou mais expressões **lógico-relacional**. Há varias formas de se categorizar as sintaxes das estruturas condicionais:
+ Simples (if);
+ Composta (if-else);
+ Aninhadas;
+ Junção (elif).
