<h2>Estruturas de Repetição</h2>
<h5>Introdução</h5>
Computadores são muito eficientes em realizar operações repetitivas.<br>

Pense no seguinte problema:
+ Imprimir "aguarde", por **cem** vezes, na tela!

Uma **estrutura de repetição** permite **repetir** um **bloco de comando** sem a necessidade de replicar código-fonte.<br>

*Observação:* **Repetição** e **iteração** tem o mesmo sigficado.<br>

Existem comandos que permitem que a máquina execute tarefas repetitivas, mais conhecidas como **laços* ou *loops**sem que seja necessário replicar o codigo.<br>
Essas estruturas de repetição sempre deve possuir um criterio de parada, caso contrario, poderá ocorrer a falha de um loop infinito, onde o bloco de comando será executado sem interrupção.<br>
Atualmente existem duas estruturas de repetição, sendo uma delas a estrutura condicional **while** e a estrutura de contagem **for**.<br>

<h5>Estrutura de Contagem - <b>FOR</b></h5>
Repete um bloco de comando por um numero finito de vezes. <br>

A garantia de repetições é controlada por uma **variável**.<br>
+ Gera-se uma sequência numérica;
+ Variavel contadora (laço contado).

<b>Sintaxe do Comando FOR</b><br>

for <'variável> in range( <'ini.>, <'fim>, <'incr./decr.>):<br>
----<'blocos do comandos'>

<b>Legenda:</b><br>

+ <variável> - Variável de controle (Contador)
+ <ini.> - Valor que será atribuido inicialmente à <variável>
+ <fim> - Valor imediatamente posterior ao último valor assumido pela <variável>
+ <incr./decr.> - Geralmente tem-se um valor de incremente ou decremento da <variável> (atualização do valor da variável)

```python
for i in range (0, 100, 1):
    print("aguarde")
```

<h5>Estrutura condicionada - <b>WHILE</b></h5>
Em muitas situações na computação precisamos repetir operações apenas enquanto uma condição for verdadeira.

+ Os comandos são repetidos <i>ate que a condição seja quebrada - condição testada no inicio</i>
+ Muitas vezes é impossivel prever quando a condição será quebrada;
+ Às vezes é inevitavel realizar repetição por contagem (comando FOR);
+ Condição: expressão lógico-relacional

<b>Sintaxe do Comando WHILE</b><br>

while <condição>:<br>
----<'blocos do comandos'>

```python
i = 0
while i < 10:
    print("aguarde")
    i += 1
```
<b><br>Estrutura de Repetição Aninhada</b></br>

Imagine uma estrutura de repetição do tipo while.
+ Considere que, dentro do bloco de comandos desta estrutura, foi inserida outra estrutura de peretição do tipo While

```python
x = 0
y = 0
while x < 100:
    x += 1
    while y < 10:
        print(f"X: {x}* Y: {y} = {x*y}")
        y += 1
    y=0

```
