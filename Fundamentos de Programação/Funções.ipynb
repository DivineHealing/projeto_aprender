<h3><strong>Funções</strong></h3>
Uma função é uma sub-rotina, um bloco de instruções que realiza tarefas específicas<br>

Durante a execução são carregadas apenas uma vez e podem ser executadas quantas vezes for necessário.<br>

São utilizadas para se dividir um problema em problemas menores (subproblemas).
+ Modulação: O Problema é dividido em partes coerentes;
+ Evisa-se repetição de código;
+ Legibilidade facilitada (abstração da complexidade).

Exemplos:<br>
+ Validação de CPF e/ou CNPJ;
+ Encontrar um CEP;
+ Calcular um frete;
+ Inicialização de vetores e matrizes;
+ Exibição de mensagens.

As sub-rotinas são muito úteis para se <ins>reutilizar uma solução</ins>, dentro do algoritmo.<br>

<b>É um sub algoritmo</b> - Possui os mesmos elementos de um algoritmo:

+ Nome;
+ Declaração de variaveis
+ Início;
+ Processo;
+ Fim.

Caracteristica especial
+ É possivel definir um <ins>valor de retorno</ins>

<b style="color:red">Sintaxe de declaração de uma função:</b>

```python
def nome_da_função(parametros):
    corpo_da_função = str()
    return valor_de_retorno
```

Parametros<br>
+ Lista de variaveis de entrada;
+ cada variável possui um nome;
+ Podem existir nenhum, um ou vários parâmetros.

**Valor de retorno:** Não é obrigatório em Python.

<h4><b>Exemplos</b></h4>
<h5>Numero 1</h5>
Faça um programa que defina uma função denominada soma() para efetuar leitura de dois números, informados diretamente pelo usuário, dentro da função. A função deve processar a soma entre os números e realizar a respectiva saída dos resultados. No corpo principal do programa invoque a função soma para que ela seja executada.

```python
def soma():
    x = int(input("Insira o valor de X: "))
    y = int(input("Insira o valor de Y: "))
    print(f"O valor da soma de X = {x} e Y = {y} foi de {x+y}")

soma()
```
<h5>Numero 2</h5>
Faça um programa que defina uma função denominada soma() que receba dois números por parâmetro. A função deve retornar a soma entre os números. No corpo principal do programa, solicite que o usuário informe os respectivos números e, então, invoque a função soma() oferecendo os dados informados pelo usuário como argumentos para os parâmetros da função. Ao final, imprima na tela o resultado retornado pela função.

```python
def soma(x, y):
    total = x + y # Variável Local
    return total

x = int(input("Insira o valor de X: ")) # Variável global
y = int(input("Insira o valor de Y: "))
    
resultado = soma(x, y)

print(f"O valor da soma de X e Y foi de {resultado}")
```

<h4><b>Escopo de Variaveis</b></h4>
O escopo de uma variável está ligado à sua visibilidade dentro de um progrma<br>

+ **Variável global** declarada no corpo principal do algoritmo, é <ins>visível em todo o programa</ins>
<br><sub>São criadas quando programa passa a executar.</sub>

+ **Varável local** declarada dentro de uma sub-rotina, <ins>só pode ser acessada dentro da sub-rotina</ins>
<br><sub>Criadas apenas quando o programa invoca a sub-rotina.</sub>
<br><sub>Quando a execução da sub-rotina acaba, a variável é destruida.</sub>

<h5>Exemplos Explicativos</h5>
Nos exemplos acima, numero 1 e 2, temos como observar as variáveis globais e locais. No caso da variavel global, temos as variáveis do exemplo de numero 2, onde X e Y estão instanciados no codigo base, assim podendo ser chamado em outras funções, enquanto dentro da função de **soma()**, temos a variável total, que so irá existir enquanto a função estiver sendo chamada.<br>

<h4><b>Recursividade</b></h4>
<h5>Definição</h5>
O conceito de recursividade está relacionado a algo que pode ser definido em termos de si próprio.<br>

+ <b>Em programação</b> Recursividade é um mecanismo que permite uma função chamar a si mesma direta ou indiretamente.
+ <b>Exemplo:</b> Relação de parentesco (Pai x Filho):
    + <b style="color:red">Pai</b>
    + <b style="color:blue">Avô</b> — <b style="color:blue">Pai</b> do meu <b style="color:red">Pai</b>
    + <b style="color:Green">Bisavô</b> — <b style="color:Green">Pai</b> do meu <b style="color:blue">Avô</b>: <b style="color:Green">Pai</b> do <b style="color:blue">Pai</b> do meu <b style="color:red">Pai</b>
    + <b style="color:Purple">Trisavô</b> — <b style="color:Purple">Pai</b> do meu <b style="color:Green">Bisavô</b>: <b style="color:Purple">Pai</b> do <b style="color:Green">Pai</b> do <b style="color:blue">Pai</b> do meu <b style="color:red">Pai</b>

O princípio da recursividade consiste em **diminuir um problema**, sucessivamente, em **problemas menores**.<br>
+ De tanto diminuir o problema, chega em um ponto que o problema pode ser resolvido **diretamente** (por definição) – Depois, combinam-se as soluções!

**A função recursiva precisa ter duas partes:**
+ **Caso base**: Determina que o problema está no menor tamanho possível (não é possivel diminuir mais).
    + Solução direta.
+ **Chamada recursiva**: Que faz uma nova chamada da função transformando o problema atual num problema menor.

<h5>Exemplo</h5>
<b> Função Fatorial</b>

Na matemática, a função fatorial de um número **n**, onde **n** ɛ **Z<sub>+</sub>**, é o resultado do produto de todos os inteiros positivos que são menores ou iguais ao próprio **n**.<br>
ou seja:
+ <ins>n! = n * (n-1) * (n-2) * ... * 3 * 2 * 1</ins><br>

Exemplo:
+ 5! = 5 * 4 * 3 * 2 * 1 = 120

Fatorial:<br>
+ 0! = 1
+ 1! = 1
+ 2! = 2 x 1 = 2
+ 3! = 3 x 2 x 1 = 6
+ 4! = 4 x 3 x 2 x 1 = 24
+ 5! = 5 x 4 x 3 x 2 x 1 = 120
+ 6! = 6 x 5 x 4 x 3 x 2 x 1 = 6 x 5! = 720

```python
def fat(n):
    if n == 1:
        return 1
    else:
        return n * fat(n-1)
    
print(fat(5))
```
