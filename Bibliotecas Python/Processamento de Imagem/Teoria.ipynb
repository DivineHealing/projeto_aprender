Por exemplo, quando se tira uma foto com o celular e se usa os filtros do aplicativo de camera para realizar melhoria ou ajuste na imagem, o que o algoritmo faz e colocar em ação funcionalidades de processamento de imagem.<br>

A ima gem digital e um quadro de pixels no formato retangular (ou quadrilatero) que representa uma séries de valores de intensidade e que são ordenados dentro de um sistema de coordenadas (x, y). De fato, trata-se apenas de um grande quadro serial de números que é interpretado por um computador para reproduzir digitalmente uma cena original.<br>

<h3>Visão computacional</h3>
<h5>Conceitos</h5>
A visão computacional é a imagem entrando para dentro do computador, sendo lida no formato de dados numericos e estão servindo a uma série de aplicações diferentes na vida social.<br>
Visão computacional é um conjunto de métodos e técnicas utilizadas para interpretar imagens, auxiliando na tomada de decisões, a partir de reconhecimento de padrões.<br>

Então, quando se fala em processamento de imagem, estamos dizendo que a entrada e a saída do processo são imagens.<br>

Realiza-se o processamento de imagem com o proposito de:
+ Melhorar a informação visual para interpretação do homem ou da máquina;
+ Melhorar as condições de armazenamento ou transmissão do arquivo da imagem;
+ Realizar efeitos digitais.

O Processamento de imagem permite um número crescente de aplicações em duas categorias distintas:
+ Aprimoramento de informação pictóricas para interpretação humana;
+ Análise automatica por computador de informações extraídas de uma cena.

<h3>OpenCV</h3>

+ Áreas de Aplicação;
+ Interface Homem-Maquina (HCI);
+ Segmentação e reconhecimento;
+ Detecção de objetos;
+ Sistema de Reconhecimento Facial;
+ Reconhecimento de gestos;
+ Estimativa de egomoção;
+ Compreensão de Movimentos
+ Rastreamento de Movimento;
+ Estrutura de Movimento (SFM);
+ Gravação de vídeos;
+ Robótica Movel;
+ Kits de ferramentas de recursos 2D e 3D;
+ Reconstrução 3D;
+ Realidade Vitual;
+ Realidade Aumentada;
+ Estereopsia visão estéro: A Percepção de profundidade a partir de 2 câmeras.

Operações Aritmeticas de Pixel a Pixel:
 - Transformações de Intensidade da Iamgem (Alteração do brilho ou do contraste);
 - Inversão da imagem (negativo).

Operações Lógicas: Transformações de união, interseção e subtração de conjuntos da imagem.

Transformações orientadas a vizinhança: Um pixel da imagem resultante depende de uma vizinhança do mesmo pixal na imagem original.<br>

**Exemplo**
 - Filtragem no dominio espacial (mascaras), mascaras de siavização(Gaussiana, Blur, *Blur* por média e mediana)
 - Transformadas de Fourier, limiarização, segmentação por regiões.
