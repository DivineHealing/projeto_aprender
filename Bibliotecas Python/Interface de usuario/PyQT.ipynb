Uma GUI (Graphical User Interface) ou interface grafica do usuário consiste em um modelo de interface responsavel pela interação entre o usuario e o sistema.<br>

Python Gui Project ou PyGui Framework é uma api que permite criarmos interfaces usando elementos nativos para aplicações em python. PyGui suporta aplicação para diversos sistemas operacionais, como Unix, Windows, Mac OS.<br>

<h5>Como criar uma GUI em Python</h5>

Ao pensarmos em desenvolvimento de software, é imprescindivel conhecermos bem o cliente, suas necessidades, as necessidades da organização, mas acima de tudo ter dominio sobre a linguagem na qual iremos desenvolver e buscar os recursos necessarios para cumprir esse projeto.

No caso de interfaces graficas, temos algumas tecnologias disponiveis.
+ O **PyQt** é um empacotador do Python para a biblioteca Qt.
+ O **PySide** é um empacotador pytrhon para biblioteca Qt. É usado como alternativa para o **Tkinter**
+ O **TKinter** é um pacote orientado a objetos que permite a criação de interfaces graficas, é identificado como padrão e pode ser usado por outras linguagens.

Outras alternativas são:
+ GTK com PyGObject e wxWidgets com Python.
+ O que todas as bibliotecas e empacotadores tem em comum são suas funcionalidade para criação de GUI's

<h5>Usando a Biblioteca PyQT</h5>

Qt: Framework multiplataforma para desenvolvimento de aplicações desktop, suportando atualmente grande aprte dos sistemas mobile.<br>
Foi aprimorado e portado para uso de linguagens como Ruby ou Java originando os bindings do Qt.<br>
No Python existem dois bindings: O **PyQt** e o **PySide**

Para criar a nossa interface grafica usaremos o **Qt Designer**, que é um editor visual que gerará um arquivo XML com a descrição da interface, que será transmitida para o codigo nativo do Python.

Instalae o PyQt5 e o QtDesigner.
